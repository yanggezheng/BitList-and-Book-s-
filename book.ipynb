```python
import pandas as pd
```

```python
import numpy as np

with open("./data/pg1513.txt") as f:
    arr = np.array(f.readlines())
Romeo_count = np.char.count(arr, "Romeo")
total_count = Romeo_count.sum()
print("Romeo is mentioned ", total_count, " times in the file.")
```

```python
Juliet_count = np.char.count(arr, "Juliet")
total_count = Juliet_count.sum()
print("Juliet is mentioned ", total_count, " times in the file.")
```

```python
import collections


with open("./data/pg1513.txt") as f:
    text = f.read()
    count = collections.Counter(text.split())
count
```

```python
sorted_count = count.most_common()
sorted_count
```

```python

stopwords = ['I', 'and', 'he', 'she', 'the', 'to', 'of', 'in', 'is', 'my', 'with', 'that', 'a','your','you','And','not', 'thou', 'be','this','for', 'me', 'it', 'thy', 'as', 'will', 'his','her', 'have', 'by', 'or', 'O', 'but']

filtered_count = [(word, count) for word, count in count.items() if word not in stopwords]
sorted_count = sorted(filtered_count, key=lambda x: x[1], reverse=True)
sorted_count
```
